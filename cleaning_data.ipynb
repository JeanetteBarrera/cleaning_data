```python
#Procesamiento
import pandas as pd
import numpy as np
import datetime as dt
#import missingno as msno

#Visualización
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
df = pd.read_csv('./datasets/android-games.csv')
df.describe().T
```

## Verificamos cuantos datos nulos hay por columna

```python
missing_values_count  = df.isna().sum()
missing_values_count
```

```python
# Filtramos para visualizar las filas que contengan datos nulos

df[df.isna().any(1)]
```

Reemplazamos los valores nulos - Nan de average rating por 4



```python
# .fillna -> encuentra los valores nulos y los reemplaza por el valor que le pasemos
# inplace -> para que modifique el dataframe original.

values = {"average rating" : 4}
df.fillna(value = values, inplace = True)
df.info()
```

Validamos los valores de la columna paid con los valores en la columna precio

```python
# Recorremos el dataframe, validamos y lo modificamos

for i in range(len(df['price'])):
    if df['price'][i] != 0:
        df['paid'][i] = True
    else:
        df['paid'][i] = False
```

```python
df.info()
```

```python
# Podemos visualizar que ya no hay valores nulos en nuestro dataframe
```

```python
df.isna().sum() # sumamos los valores nulos en cada columna
```

### Verificamos si existen valores duplicados

```python
df[df.duplicated()]
```

```python
df[df.duplicated(subset=['Date','rank','title','total ratings','category'], keep = False)]
```

Podemos observar que la diferencia entre los datos duplicados se encuentra en las variables.
- 5 star ratings
- 4 star ratings
- 3 star ratings
- 2 star ratings
- 1 star ratings

Para eliminarlos y quedar con uno de los datos podemos tomar el promedio de las variables mencionadas.

```python
top_games = df.groupby(['Date','rank','title','total ratings','installs','average rating',
                        'growth (30 days)','growth (60 days)','price','category','paid'])['5 star ratings','4 star ratings',
                        '3 star ratings','2 star ratings','1 star ratings'].mean()

top_games.reset_index(inplace=True)
```

```python
top_games[top_games.duplicated(subset=['Date','rank','title','total ratings','category'], keep = False)]
```

```python
top_games.sample(5)
```

### Uniformedad de datos categoricos

```python
top_games['category'] = top_games['category'].str.upper() 
# transforma todas las categorias a mayusculas
```

```python
top_games['category'].unique()
```

Podemos observar categorias que no corresponden videojuegos, esas categorias corresponden a libros
En este caso podemos eliminarlos

```python
top_games = top_games[~top_games['category'].isin(['FICTION BOOK', 'BIOGRAPHY BOOK'])]
```

```python
# Verificamos que ya no se encuentran estas categorias en nuetro dataframe

top_games['category'].unique()
```

## Tipos de datos

La columna de precio es de tipo object, esta columna nos será más útil si lo pasamos a tipo int o float para futuros calculos y gráficas:

```python
def installs(x):
    if x[-1] == 'M':
        return(float(x[:-2])*1000000)
    else:
        return(float(x[:-2])*1000)
```

```python
top_games['installs'] = top_games['installs'].apply(installs)
```

```python
# Verificamos

display(top_games['installs'].unique())
display(top_games['installs'].dtype)
```

```python
top_games['installs'].astype('int')
```

## Reemplazar valores de variables

Vamos a cambiar los valores de la columna que indica si el videojuegos es pago o gratis:

- Pago (True) = 1
- Gratis (False) = 0

```python
top_games['paid'].unique()
```

```python
top_games['paid'] = top_games['paid'].replace({True:1, False:0})
```

```python
top_games['paid'].unique()
```

## Formato de fechas

Validar el formato de la fecha y convertirlo a datetime:

```python
# Unificamos el formato

top_games['Date'] = pd.to_datetime(top_games['Date']).dt.strftime('%m/%d/%Y')
```

```python
#Validamos

display(top_games['Date'].unique())
display(len(top_games['Date'].unique()))
```
